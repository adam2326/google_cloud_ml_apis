```python
# write to cloud storage
sample_bucket = storage.Bucket("machine-learning-backend-storagebucket")
leftovers= list()
for image in images:
  try:    
    IMAGE="https://storage.googleapis.com/machine-learning-backend-storagebucket/" + image
    vservice = build('vision', 'v1', developerKey=APIKEY)
    request = vservice.images().annotate(body={
            'requests': [{
                    'image': {
                        'source': {
                            'imageUri': IMAGE
                        }
                    },
                    'features': [{
                        'type': 'DOCUMENT_TEXT_DETECTION',
                        'maxResults': 1,
                    }]
                }],
            })
    responses = request.execute(num_retries=3)
    text_out = responses['responses'][0]['fullTextAnnotation']['text'].encode('ascii','ignore')
    sample_object = sample_bucket.object('sas_data/HLS/output/' + image.split('/')[-1] + '.txt')
    sample_object.write_stream(text_out, 'text/plain')
  except:
    #responses['responses'][0]['fullTextAnnotation']['text'][:50]
    leftovers.append(image)
```
